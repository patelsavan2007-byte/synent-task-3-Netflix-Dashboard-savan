```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df['director'].fillna('Unknown',inplace=True)
df['country'].fillna('Unknown',inplace=True)
df['date_added'].fillna('Unknown',inplace=True)
```

```python
df['cast'].fillna('Not Available',inplace=True)
df['duration'].fillna('Not Available',inplace=True)
df['rating'].fillna(df['rating'].mode()[0],inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='type',data=df)
plt.title('Movies vs Tv-Shows')
plt.show()
```

```python
type_counts = df['type'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(type_counts,labels=type_counts.index,autopct='%1.1f%%')
plt.title("Content Type Distribution")
plt.show()
```

# Content Added Over Years
#### Insight:-
#### Huge growth after 2015
#### Peak content additions around 2019–2020

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='year_added',data=df,order=df['year_added'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Content Added each year')
plt.show()
```

## Top Directors

```python
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_directors.values,
            y=top_directors.index)

plt.title("Top Directors on Netflix")
plt.show()
```

# Top Genres on Netflix

```python
geners = df['listed_in'].str.split(',',expand=True)
all_genres = pd.Series(geners.values.ravel())
all_genres = all_genres.str.strip()
top_genres = all_genres.value_counts().head(10)
plt.figure(figsize=(10,5))

sns.barplot(x=top_genres.values,
            y=top_genres.index)

plt.title("Top Genres on Netflix")
plt.show()
```

## Create Duration Numeric Feature & Correlation Matrix

```python
# Extract numeric duration
df['duration_num'] = df['duration'].str.extract('(\d+)')

df['duration_num'] = pd.to_numeric(df['duration_num'], errors='coerce')
corr = df[['release_year', 'duration_num']].corr()

print(corr)
```

```python
plt.figure(figsize=(5,4))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```
